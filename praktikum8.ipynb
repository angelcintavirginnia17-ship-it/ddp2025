```python
print ("mengkonversi suhu dari celcius ke fahrenheit")
def celcius_ke_fahrenheit(celcius):
    fahrenheit = (celcius * 9/5) + 32
    return fahrenheit

print(celcius_ke_fahrenheit(0))
print(celcius_ke_fahrenheit(100))
```

```python
print("mengembalikan true jika bilangan genap, false jika bilangan ganjil")
def is_genap(bilangan_bulat):
    if bilangan_bulat % 2 == 0:
        return True
    else:
        return False


print(is_genap(4))
print(is_genap(7))
```

```python
print(" menampikan bilangan ganjil kurang dari batas_atas.")
def bilangan(batas_atas):
    ganjil = []
    for i in range(1, batas_atas):
        if i % 2 != 0:
            ganjil.append(i)

    print(*ganjil, sep=",")
bilangan(20)
```

```python
print("menampilkan keterangan 'LULUS' atau 'GAGAL' berdasarkan skor.")
def nilai(skor):
    batas_lulus = 70
    if skor >= batas_lulus:
        print("LULUS")
    else:
        print("GAGAL")


nilai(80)
nilai(60)
```
